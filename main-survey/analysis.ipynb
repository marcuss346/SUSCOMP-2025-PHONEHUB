```python
def count_gender_faculty_table(data_str):
    lines = data_str.strip().split("\n")

    faculty_totals = {}
    faculty_gender = {}

    for line in lines:
        # split only on the first tab
        gender, faculty = line.split("\t", 1)

        gender = gender.strip()
        faculty = faculty.strip()

        faculty_totals[faculty] = faculty_totals.get(faculty, 0) + 1

        key = (faculty, gender)
        faculty_gender[key] = faculty_gender.get(key, 0) + 1

    genders = ["Moki", "enska", "Ne elim odgovoriti."]

    print("\\begin{tabular}{lrrrrr}")
    print("\\hline")
    print("Faculty & Total & Moki & enska & Ne elim odgovoriti & Other \\\\")
    print("\\hline")

    for faculty, total in faculty_totals.items():
        row = [faculty, total]

        other_count = total
        for g in genders:
            count = faculty_gender.get((faculty, g), 0)
            row.append(count)
            other_count -= count

        row.append(other_count)

        print(f"{row[0]} & {row[1]} & {row[2]} & {row[3]} & {row[4]} & {row[5]} \\\\")

    print("\\hline")
    print("\\end{tabular}")
```

```python
import re
from collections import defaultdict

def select_label(faculty):
    if faculty == "Fakulteta za matematiko in fiziko":
        return "FMF"
    elif faculty == "Fakulteta za raunalnitvo in informatiko":
        return "FRI"

def count_age_faculty_hist(data_str):
    # helper to normalize ages to integers
    def parse_age(raw):
        raw = raw.strip()
        match = re.match(r"(\d+)", raw)
        if match:
            return int(match.group(1))
        return None

    counts = defaultdict(lambda: defaultdict(int))

    for line in data_str.strip().split("\n"):
        parts = line.split("\t")
        if len(parts) < 3:
            continue
        
        age_raw, _, faculty = parts[0], parts[1], parts[2]
        age = parse_age(age_raw)

        if age is None:
            continue  # skip unparseable ages

        counts[age][faculty] += 1

    # ---- prepare plotting data ----
    ages = sorted(counts.keys())
    faculties = sorted({f for a in counts for f in counts[a]})

    # build bars
    data_by_faculty = {f: [counts[a].get(f, 0) for a in ages] for f in faculties}

    # ---- stacked histogram (two-tone) ----
    bottom = [0] * len(ages)

    for faculty in faculties:
        plt.bar(ages, data_by_faculty[faculty], bottom=bottom, label=select_label(faculty))
        bottom = [b + v for b, v in zip(bottom, data_by_faculty[faculty])]

    plt.xlabel("Age")
    plt.ylabel("Number of students")
    plt.title("Age Distribution by Faculty")
    plt.legend()
    #plt.tight_layout()
    plt.grid(axis='y', color='#ddddee')

    plt.xticks(ages)

    plt.show()

    # return the data before plotting (as requested)
    return counts
```
