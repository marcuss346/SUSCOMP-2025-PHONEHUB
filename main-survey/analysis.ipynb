```python
def count_gender_faculty_table(data_str):
    lines = data_str.strip().split("\n")

    faculty_totals = {}
    faculty_gender = {}

    for line in lines:
        # split only on the first tab
        gender, faculty = line.split("\t", 1)

        gender = gender.strip()
        faculty = faculty.strip()

        faculty_totals[faculty] = faculty_totals.get(faculty, 0) + 1

        key = (faculty, gender)
        faculty_gender[key] = faculty_gender.get(key, 0) + 1

    genders = ["Moki", "enska", "Ne elim odgovoriti."]

    print("\\begin{tabular}{lrrrrr}")
    print("\\hline")
    print("Faculty & Total & Moki & enska & Ne elim odgovoriti & Other \\\\")
    print("\\hline")

    for faculty, total in faculty_totals.items():
        row = [faculty, total]

        other_count = total
        for g in genders:
            count = faculty_gender.get((faculty, g), 0)
            row.append(count)
            other_count -= count

        row.append(other_count)

        print(f"{row[0]} & {row[1]} & {row[2]} & {row[3]} & {row[4]} & {row[5]} \\\\")

    print("\\hline")
    print("\\end{tabular}")
```

```python
import re
from collections import defaultdict

def select_label(faculty):
    if faculty == "Fakulteta za matematiko in fiziko":
        return "FMF"
    elif faculty == "Fakulteta za raunalnitvo in informatiko":
        return "FRI"

def count_age_faculty_hist(data_str):
    # helper to normalize ages to integers
    def parse_age(raw):
        raw = raw.strip()
        match = re.match(r"(\d+)", raw)
        if match:
            return int(match.group(1))
        return None

    counts = defaultdict(lambda: defaultdict(int))

    for line in data_str.strip().split("\n"):
        parts = line.split("\t")
        if len(parts) < 3:
            continue
        
        age_raw, _, faculty = parts[0], parts[1], parts[2]
        age = parse_age(age_raw)

        if age is None:
            continue  # skip unparseable ages

        counts[age][faculty] += 1

    # ---- prepare plotting data ----
    ages = sorted(counts.keys())
    faculties = sorted({f for a in counts for f in counts[a]})

    # build bars
    data_by_faculty = {f: [counts[a].get(f, 0) for a in ages] for f in faculties}

    # ---- stacked histogram (two-tone) ----
    bottom = [0] * len(ages)

    for faculty in faculties:
        plt.bar(ages, data_by_faculty[faculty], bottom=bottom, label=select_label(faculty))
        bottom = [b + v for b, v in zip(bottom, data_by_faculty[faculty])]

    plt.xlabel("Age")
    plt.ylabel("Number of students")
    plt.title("Age Distribution by Faculty")
    plt.legend()
    #plt.tight_layout()
    plt.grid(axis='y', color='#ddddee')

    plt.xticks(ages)

    plt.show()

    # return the data before plotting (as requested)
    return counts
```

```python
def extract_column_by_number(csv_path, column_index):
    """
    Extracts a column from a CSV file by column index (0-based).
    Row 0 = header, data starts from row 1.

    Returns:
        original_list
    """

    original_values = []

    with open(csv_path, encoding="cp1250") as f:
        reader = csv.reader(f)
        rows = list(reader)

        # Validate index
        if column_index >= len(rows[0]):
            raise IndexError(
                f"Column index {column_index} out of range. "
                f"CSV has {len(rows[0])} columns."
            )

        # Extract data starting from row 1
        for row in rows[1:]:
            value = row[column_index].strip()
            original_values.append(value)

    # Print results
    print("\n=== COLUMN DATA ===")
    print(f"Column index: {column_index}")
    print(original_values)

    return original_values
```

```python
faculty_list = extract_column_by_number(file_name, 8)
change_reason_list = extract_column_by_number(file_name, 14)
```

```python
from collections import defaultdict, Counter

def normalize_faculty(name: str) -> str:
    """Fix encoding artifacts and unify faculty names."""
    name = name.lower()
    name = name.replace("?", "")
    name = name.replace("ra?unal", "raunal")
    name = name.replace("matematiko informatiko", "matematiko in fiziko")
    return name.strip()

def analyze_faculty_vs_reason(faculty_list, reason_list):
    """
    Compare two survey columns:
      - faculty_list: column 8
      - reason_list: column 14

    Produces:
      - total counts per faculty
      - total counts per reason
      - cross-tabulation: reason frequency per faculty
    """

    # Normalize faculty names
    faculty_norm = [normalize_faculty(f) for f in faculty_list]

    # Identify faculty groups
    FMF = "fakulteta za matematiko in fiziko"
    FRI = "fakulteta za raunalnitvo in informatiko"

    # Prepare containers
    faculty_counts = Counter(faculty_norm)
    reason_counts = Counter(reason_list)

    cross_tab = defaultdict(lambda: Counter())

    # Build cross-tabulation
    for fac, reason in zip(faculty_norm, reason_list):
        cross_tab[fac][reason] += 1

    # Print results
    print("\n=== TOTAL RESPONSES PER FACULTY ===")
    for fac, count in faculty_counts.items():
        print(f"{fac}: {count}")

    print("\n=== TOTAL RESPONSES PER REASON ===")
    for reason, count in reason_counts.items():
        print(f"{reason}: {count}")

    print("\n=== CROSS-TABULATION (FACULTY  REASON) ===")
    for fac, reasons in cross_tab.items():
        print(f"\n--- {fac.upper()} ---")
        for reason, count in reasons.items():
            print(f"{reason}: {count}")

    return {
        "faculty_counts": faculty_counts,
        "reason_counts": reason_counts,
        "cross_tab": cross_tab
    }
```

```python
import csv

def normalize_text(s: str) -> str:
    """Fix encoding artifacts and unify Slovenian characters."""
    if not isinstance(s, str):
        return ""
    s = s.lower()
    s = s.replace("?", "")
    s = s.replace("ra?unal", "raunal")
    return s.strip()

def separate_fmf_fri(csv_path, encoding="cp1250"):
    """
    Reads a CSV and separates rows into:
      - FMF (Fakulteta za matematiko in fiziko)
      - FRI (Fakulteta za raunalnitvo in informatiko + all exceptions)

    Returns:
        fmf_rows (list of dicts)
        fri_rows (list of dicts)
    """

    FMF = "Fakulteta za matematiko in fiziko"
    FRI = "Fakulteta za raunalnitvo in informatiko"

    fmf_rows = []
    fri_rows = []

    with open(csv_path, encoding=encoding) as f:
        reader = csv.DictReader(f)

        for row in reader:
            faculty_raw = row.get("Na kateri fakulteti tudirate?", "")
            faculty = normalize_text(faculty_raw)

            if FMF in faculty:
                fmf_rows.append(row)
            else:
                # Everything else goes to FRI group
                fri_rows.append(row)

    # Print results
    print("\n=== FMF ROWS ===")
    for r in fmf_rows:
        print(r)

    print("\n=== FRI + EXCEPTIONS ROWS ===")
    for r in fri_rows:
        print(r)

    return fmf_rows, fri_rows
```
