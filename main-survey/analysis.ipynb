## Demography

```python
#Koliko ste stari?
ages = [22, 19, 20, 22, 22, 22, 23, 23, 19, 22, 21, 21, 21, 20, 24, 20, 21, 22, 23, 19, 21, 21, 23, 19, 21, 24, 23, 19, 23, 19, 22, 21, 22, 25, 23, 19, 20, 21, 22, 21, 20, 24, 22, 22, 22, 19, 24, 22, 23, 20, 22, 23, 20, 20, 21, 28, 20, 22, 23, 19, 21, 25, 23, 20, 21, 21, 25, 21, 21, 26, 23, 26, 22, 21, 26, 21, 21, 19, 23, 21, 19, 21, 20, 22, 19, 22, 23, 19, 19, 20, 27, 22, 20, 20, 22, 21, 19, 24, 20, 22, 27, 22, 21, 20, 19, 24, 21, 22, 20, 20, 20, 22, 21, 22]
counter = {}
for age in ages:
    if age not in counter:
        counter[age] = 1
    else:
        counter[age] += 1

from collections import OrderedDict
sorted_counter = OrderedDict(sorted(counter.items()))
print(sorted_counter)

print("Age distribution:")
for age, count in sorted_counter.items():
    print(f"Age {age}: {count}")

```

```python
#Make age histogram
import matplotlib.pyplot as plt

plt.bar(sorted_counter.keys(), sorted_counter.values())
plt.xlabel('Age')
plt.ylabel('Number of People')
plt.title('Age Distribution Histogram')
plt.xticks(list(sorted_counter.keys()))
plt.show()
```

## Personal Mobile Device

```python
brands = {"Samsung" : 48, "Apple" : 25, "Xiaomi" : 20, "Google" : 7, "Oneplus" : 1, "Oppo" : 1, "Realme" : 2, "Fairphone" : 1, "Huawei" : 2, "Motorola" : 3, "Poco" : 1, "Sony" : 1, "Nothing" : 2}



# Key, Value of the dictionary defined
#d = {'watermelon': 1, 'apple': 2, 'banana': 3}        

# Sort based on Values
val_based = {k: v for k, v in sorted(brands.items(), key=lambda item: item[1])}
# item[1] represents the sorting based on value

# Sort based on reverse of Values
val_based_rev = {k: v for k, v in sorted(brands.items(), key=lambda item: item[1], reverse=True)}

# Print sorted dictionary
print(val_based)
print(val_based_rev)


```

```python
plt.bar(val_based_rev.keys(), val_based_rev.values())
plt.xlabel('Phone Brand')
plt.ylabel('Number of People')
plt.title('Phone Brand Distribution Histogram')
plt.xticks(list(val_based_rev.keys()), rotation='vertical')
plt.show()
```
