```python
def calculate_percentage(data_dict):
    total = sum(data_dict.values())
    percentage_dict = {key: int((value / total) * 100) for key, value in data_dict.items()}
    print("Percentage distribution:")
    for key, percentage in percentage_dict.items():
        print(f"{key}: {percentage:.0f}%")
    return percentage_dict
```

### Phone change reason analysis

From the survey results it is quite clear that a large part of participants felt the need to change their previous phone due to the following reasons (with percents that show what part of the whole group of participants chose this as the most possible reason): "Irreparable damage or damage that is too expensive to repair." (38 %),  "Slow performance / obsolescence." (27 %), "Poor battery." (15 %), "I want a newer model" (4 %). 13 % of all participants that answered this questions stated their own reason, out of these the majority mentioned that their reason for the change (in one way or another) included the following: <br>
- exchanging phones with parents (either parents give the phone to the student or the student gives the phone to a parent)
- the new phone was recieved as a gift / for free and the old phone lost its use
The other participants wrote out some form of damage to the phone or absolescence of the device or OS.
With **Other** the pie chart includes reasons that were chosed less frequently such as "Security/system updates are no longer available.", "Running out of space." and "Favorable offer from the operator".

```python
plt.bar(phone_change_reasons.keys(), phone_change_reasons.values(), color="green")
plt.xlabel('Phone Change Reasons')
plt.ylabel('Number of Students')
plt.title('Phone Change Reasons Distribution Histogram')
plt.xticks(list(phone_change_reasons.keys()), rotation='vertical')
plt.show()
```

```python
y1 = list(phone_change_reasons.values())
labels1 = list(phone_change_reasons.keys())

title = plt.title("Phone Change Reasons Pie Chart")
title.set_ha("left")
plt.gca().axis("equal")
pie = plt.pie(y1, autopct='%1.0f%%', startangle=0)
plt.legend(pie[0],labels1, bbox_to_anchor=(1.3,0.5), loc="center right", fontsize=10, 
           bbox_transform=plt.gcf().transFigure)
plt.subplots_adjust(left=0.0, bottom=0.1, right=0.5)
```
